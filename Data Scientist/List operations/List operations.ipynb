### Parsing The File

```python
weather_data = []

f = open('la_weather.csv','r')
d = f.read()
rows = d.split('\n')

for row in rows:
    item = row.split(',')
    weather_data.append(item)
weather_data
```

### Getting A Single Column From The Data


```python
weather = []
for row in weather_data:
    weather.append(row[1])
weather
```
